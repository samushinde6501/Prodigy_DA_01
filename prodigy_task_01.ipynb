# Create a bar chart or histogram to visualize the distribution of a categorized or continuous variable,such as the distribution of ages or gender in a population.

```python
#libraries
import matplotlib.pyplot as plt   
import numpy as np
```

```python
data = {
    'Gender': ['Male', 'Female', 'Other'],
    'Count': [500, 450, 50]
}
```

```python
# Extracting data for the bar chart
categories = data['Gender']
values = data['Count']
```

```python
# Creating the bar chart
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color=['blue', 'pink', 'purple'])
# Adding title and labels
plt.title('Gender Distribution in Population')
plt.xlabel('Gender')
plt.ylabel('Count')
# Displaying the bar chart
plt.show()
```

```python
np.random.seed(0)
ages = np.random.randint(0, 100, size=1000)
```

```python
# Creating the histogram
plt.figure(figsize=(8, 6))
plt.hist(ages, bins=20, color='skyblue', edgecolor='black')
# Adding title and labels
plt.title('Age Distribution in Population')
plt.xlabel('Age')
plt.ylabel('Frequency')
# Displaying the histogram
plt.show()
```
